```python
# Vector addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

vector_sum = a + b
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"a + b: {vector_sum}")
# Geometric interpretation in 2D
a_2d = np.array([2 ,3])
b_2d = np.array([1, 2])
sum_2d = a_2d +b_2d

plt.figure(figsize=(8,6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles = "xy", scale_units = "xy", scale = 1, color = "red", width = 0.005)
plt.quiver(0, 0, b_2d[0], b_2d[1], angles = "xy", scale_units = "xy", scale = 1, color = "blue", width = 0.005)
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles = "xy", scale_units = "xy", scale = 1, color = "green", width = 0.005)
plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles = "xy", scale_units = "xy", scale = 1, color = "blue", width = 0.005, alpha = 0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule")
plt.legend(['Vector a', 'Vector b', 'a + b'])
plt.show()
```

```python
# Properties of vector addition
c = np.array([1, -1, 2])

# Commutative property: a + b = b + a
print(f"a + b = {a + b}")
print(f"b + a = {b + a}")
print(f"Commutative: {np.array_equal(a + b, b + a)}")

# Associative property: (a + b) + c = a + (b + c)
left = (a + b) + c
right = a + (b + c)
print(f"(a + b) + c = {left}")
print(f" a (b + c)= {right}")
print(f"Associative: {np.array_equal(left, right)}")

```

```python
# Scalar multiplication
vector = np.array([2, 3, 1])
scalar = 3

scaled_vector = scalar * vector
print(f"Original vector: {vector}")
print(f"Scaler: {scalar}")
print(f"Scaled vector: {scaled_vector}")

# Geometric interpretation
original_2d = np.array([2, 1])
scales = [0.5, 1, 2, -1]


plt.figure(figsize = (10, 6))
colors = ['red', 'blue', 'green', 'purple']
for i, scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0, 0, scaled[0], scaled[0], angles = "xy", scale_units ="xy", scale= 1, color=colors[i], width = 0.005, label= f"{scale} * v")

plt.xlim(-3, 5)
plt.ylim(-3, 3)
plt.grid(True)
plt.title("Scalar Multiplication")
plt.legend()
plt.show()
```

```python
# Properties of scalar multiplication
a = np.array([1, 2, 3])
c1, c2 = 2, 3

# Associative: c1(c2*a) = (c1*c2)
left_1 = c1 * (c2 * a)
right_1 = (c1 * c2) * a
print(f"c1(c2*a) = {left_1}")
print(f"(c1*c2)*a = {right_1}")
print(f"Associative: {np.array_equal(left_1, right_1)}")

# Distributive over vector addition: c(a + b) = ca + cb
left_2 = c1 * (a + b)
right_2 = c1 * a + c1 * b
print(f"c(a + b) = {left_2}")
print(f"ca + cb = {right_2}")
print(f"Distributive: {np.array_equal(left_2, right_2)}")


```

```python
# Physicist's view: vectors as arrows in space
vector_a = np.array([3, 4])
magnitude = np.linalg.norm(vector_a)
print(f"Vector: {vector_a}")
print(f"Magnitude: {magnitude}")

# Direction (unit vector)
unit_vector = vector_a / magnitude
print(f"Unit vector (direction): {unit_vector}")

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0,0, vector_a[0],vector_a[1], angles = "xy", scale_units = "xy", scale = 1, color = "red", width=0.005)
plt.quiver(0,0, unit_vector[0], unit_vector[1], angles = "xy", scale_units = "xy", scale = 1, color = "blue", width=0.003)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.title("Vector: Magnitude and Direction")
plt.legend(['Original Vector', 'Unit Vector'])
plt.show()


```

```python
# Mathematician's view: vectors as abstract objects
# Example 1: Polynomials as vectors
# p(x) = 3x^2 + 2x + 1 represented as [1, 2, 3] (coefficients)
poly1 = np.array([1, 2, 3])  # 1 + 2x + 3x^2
poly2 =np.array([2, -1, 1]) # 2 - x + x^2

# Vector addition for polynomials
poly_sum = poly1 + poly2
print("Polynomial 1:", poly1)
print("Polynomial 2:", poly2)
print("Sum:", poly_sum) # (1+2) + (2-1)x + (3+1)x^2 = 3 + x + 4x^2

# Scalar multiplication for polynomials
scalar = 2
poly_scaled = scalar * poly1
print(f"2 * Polynomial 1:", poly_scaled)

# Example 2: Functions as vectors (function space)
# This is more abstract, but functions can be added and scaled
def func1(x):
    return x ** 2
def func2(x):
    return 2*x

# Function addition: (f + g)(x) = f(x) + g(x)
# Function addition: (f + g)(x) = f(x) + g(x)
def func_sum(x):
    return func1(x) + func2(x)


x_vals = np.linspace(0, 3, 100)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, func1(x_vals), label='f(x) = x')
plt.plot(x_vals, func2(x_vals), label='g(x) = 2x')
plt.plot(x_vals, func_sum(x_vals), label='(f+g)(x) = x + 2x')
plt.legend()
plt.title("Functions as Vectors")
plt.grid(True)
plt.show()




```

```python
# Audio signals as vectors
sample_rate = 44100
duration = 0.1
t = np.linspace(0, duration, int(sample_rate * duration))
t

# Two sine waves (audio signals)
signal1 = 0.5 * np.sin(2 * np.pi * 440 * t)
signal2 = 0.5 * np.sin(2 * np.pi * 880 * t)

# Vector addition of audio signals
mixed_signal = signal1 + signal2

# Scalar multiplication (volume control)
quiet_signal = 0.5 * mixed_signal

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(t[:1000],signal1[:1000])
plt.title("Signal 1 (440 Hz)")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t[:1000],signal2[:1000])
plt.title("Signal 2 (880 Hz)")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t[:1000], mixed_signal[:1000])
plt.title("Mixed Signal (Addition)")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t[:1000], quiet_signal[:1000])
plt.title("Quiet Signal (Scalar Multiplication)")
plt.grid(True)

plt.tight_layout()
plt.show()


print(f"Signal 1 shape: {signal1.shape}")
print(f"Signal 2 shape: {signal2.shape}")
print(f"Mixed signal shape: {mixed_signal.shape}")

```

```python
# Matrix addition - element-wise operation
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6], 
              [7, 8]])

# Addition
C = A + B
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nA + B:")
print(C)
```

```python

# Properties of matrix addition
# Commutative: A + B = B + A
print(f"\nCommutative property:")
print(f"A + B = B + A: {np.array_equal(A + B, B + A)}")
```

```python
# Associative: (A + B) + C = A + (B + C)
C_new = np.array([[1, 1], [1, 1]])
left = (A + B) + C_new
right = A + (B + C_new)
print(f"Associative property:")
print(f"(A + B) + C = A + (B + C): {np.array_equal(left, right)}")
```

```python
zero_matrix = np.zeros_like(A)
print(f"\nAdditive identity:")
print(f"A + 0 = A: {np.array_equal(A + zero_matrix, A)}")
```
