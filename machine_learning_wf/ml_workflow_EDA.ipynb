| Numeric | Category   | Meaning               |

| 3–4     | **Low**    | Poor quality wines    |
| 5–6     | **Medium** | Average/typical wines |
| 7–8+    | **High**   | Premium-quality wines |


```python
# let's make a copy our dataset
df_copy = white_wine.copy()
```

```python
def map_quality_category(x):
    if x <= 4:
        return "Low"
    elif x <=6:
        return "Medium"
    else:
        return "High"


# let's apply our function
df_copy["quality_label"] = df_copy["quality"].apply(map_quality_category)
    
```
