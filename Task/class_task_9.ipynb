Task:
Is the beef consumption in Argentina significantly different from that in Bangladesh?

H₀: Mean beef consumption (Argentina) = Mean beef consumption (Bangladesh)

H₁: Mean beef consumption (Argentina) ≠ Mean beef consumption (Bangladesh)

```python
# import necessary libaries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# Select Argentina beef consumption
arg_beef = df[(df["country"]=="Argentina") & (df["food_category"]=="beef")]
```

```python
# Select Bangladesh beef consumption
bang_beef = df[(df["country"]=="Bangladesh") & (df["food_category"]=="beef")]
```

```python
arg_beef
```

```python
bang_beef
```

```python
arg_consump = arg_beef["consumption"]
arg_consump
```

```python
bang_consump = bang_beef["consumption"]
bang_consump
```

```python
arg_consump_mean = arg_consump.mean()
arg_consump_mean
```

```python
bang_consump_mean = bang_consump.mean()
bang_consump_mean
```

```python
print(f"Mean Beef Consumption in Argentina: {arg_consump_mean}")
print(f"Mean Beef Consumption in Bangladesh: {bang_consump_mean}")
```

```python
arg_beef_samples = np.random.normal(loc=arg_consump_mean, scale=5, size=30)
bang_beef_samples = np.random.normal(loc=bang_consump_mean, scale=5, size=30)
```

```python
arg_beef_samples
```

```python
bang_beef_samples
```

```python
from scipy import stats
```

```python
t_statistic, p_value = stats.ttest_ind(arg_beef_samples, bang_beef_samples)
```

```python
print("T-statistic:", t_statistic)
```

```python
print("P-value:", p_value)
```

Interpretation for Students:

If p < 0.05, reject H₀ → Mean beef consumption (Argentina) = Mean beef consumption (Bangladesh)

If p ≥ 0.05, fail to reject H₀.

Since our P-Value is less than 0.05, we reject H₀, meaning the beef consumption in Argentina is significantly different.
