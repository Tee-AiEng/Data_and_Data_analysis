```python
df["Total_Rev"] = df["Price"] * df["Units"]
df
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month_name()
df
```

## Total Units by Product

```python
product_by_units = df.groupby(["Products"])["Units"].sum().sort_values(ascending=False)
product_by_units
```

## Total Revenue by Product

```python
product_by_rev = df.groupby(["Products"])["Total_Rev"].sum().sort_values(ascending=False)
product_by_rev
```

## Sales Agent by Revenue

```python
HIgest_rev_by_sales_agent =df.groupby(["Sales Agent"])["Total_Rev"].sum().sort_values(ascending=False)
HIgest_rev_by_sales_agent
```

## Total Units by Branch

```python
Branch_by_unit =df.groupby(["Branch"])["Units"].sum().sort_values(ascending=False)
Branch_by_unit
```

## Total Revnue by Branch

```python
Branch_by_rev =df.groupby(["Branch"])["Total_Rev"].sum().sort_values(ascending=False)
Branch_by_rev
```

## Monthly Trends

```python
monthly_trends = pd.pivot_table(data = df, values="Total_Rev", columns="Year",index="Month", aggfunc= "sum")
monthly_trends
```
